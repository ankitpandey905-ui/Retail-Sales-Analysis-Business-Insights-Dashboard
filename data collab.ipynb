```
import matplotlib.pyplot as plt

plt.hist(df['Sales'], bins=80)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')


plt.show()
```

# The sales distribution is positively skewed, indicating that most transactions are of low value, while a few high-value transactions act as **outliers**

```
plt.scatter(range(len(df)), df['Sales'])
plt.title('Sales Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.show()
```

```
region_sales = df.groupby('Region')['Sales'].sum()
```

```
print(region_sales)
```

```
print("Highest Sales Region:", region_sales.idxmax())
```

```
import matplotlib.pyplot as plt

region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
colors = ['red', 'blue', 'green', 'orange']

region_sales.plot(kind='bar', color=colors)

plt.show()
```

```
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Regional Sales Share')
plt.ylabel('')
plt.show()
```

# **The region with the highest sales can be identified by grouping the dataset by Region and summing the Sales values.Western region has highest sales





```
# Group by Category and sum Sales
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)
```

```
print("Top Category:", category_sales.idxmax())
```

```
import matplotlib.pyplot as plt

plt.bar(category_sales.index, category_sales.values, color='magenta')
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

By grouping the dataset by product category and summing the sales values, we can identify the category generating the highest revenue. The category with the maximum total sales is the top-performing category,that is TECHNOLOGY

```
category_profit = df.groupby('Category')['Profit'].sum()

print(category_profit)
```

```
import matplotlib.pyplot as plt

plt.bar(category_profit.index, category_profit.values, color='orange')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()
```

By grouping the dataset by category and summing the profit values, we can identify the most profitable category. The category with the highest total profit is considered the best-performing in terms of profitability.
Technology â highest profit
